```python
import numpy as np
import PIL
from fastai.vision.all import *
from collections import Counter

# --- 1. Configure o caminho ---
path_masks = Path(r'/home/tempestade/Ãrea de Trabalho/git/geoeyeai/cell_data/masks')
arquivos = get_image_files(path_masks)

print(f"ðŸ“‚ Analisando {len(arquivos)} mÃ¡scaras... aguarde.")

# Contadores para armazenar as cores encontradas
cores_encontradas = Counter()
tipos_arquivos = Counter() # Para ver se Ã© misturado RGB e Cinza

for arquivo in arquivos:
    # Abre a imagem
    img = PIL.Image.open(arquivo)
    arr = np.array(img)
    
    # Verifica se Ã© Grayscale (2D) ou RGB (3D)
    if len(arr.shape) == 2:
        tipos_arquivos['Escala de Cinza (2D)'] += 1
        # Pega valores Ãºnicos
        uniques = np.unique(arr)
        cores_encontradas.update(uniques)
        
    elif len(arr.shape) == 3:
        tipos_arquivos['RGB (3D)'] += 1
        # Transforma a matriz 3D em uma lista de pixels (R, G, B)
        # Ex: transforma (1024, 1024, 3) em (1048576, 3)
        pixels = arr.reshape(-1, 3)
        # Pega as linhas Ãºnicas (cores Ãºnicas)
        uniques = np.unique(pixels, axis=0)
        # Adiciona ao contador (convertendo para tupla para poder contar)
        cores_encontradas.update([tuple(p) for p in uniques])

print("-" * 40)
print("ðŸ“Š RESUMO DO DIAGNÃ“STICO")
print("-" * 40)
print(f"Tipos de arquivos encontrados: {dict(tipos_arquivos)}")
print("\nðŸŽ¨ AS 10 CORES MAIS COMUNS (Formato: Cor -> Quantas imagens tÃªm essa cor):")

# Mostra as 15 cores mais frequentes
for cor, contagem in cores_encontradas.most_common(15):
    print(f"   ðŸ‘‰ Cor: {cor} \t(Aparece em pixels de vÃ¡rias imagens)")

print("-" * 40)
```
